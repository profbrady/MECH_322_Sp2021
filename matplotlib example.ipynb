## Imports

## Create data points for plotting

Use NumPy to create an array of $x$ values and then use $x$ to calculate $y$ values for plotting

```python
x = np.linspace(-np.pi*2, np.pi*2, 200)
y1 = x*np.cos(x)
y2 = x*np.sin(x)
```

## Matplotlib object-oriented plotting

Following are a number of common plotting commands that use Matplotlib's object-oriented approach. The same plot data is used a number of times with changes to the plot settings each time so the user can see what affect each change has on the plot.

### Simple Plot

```python
rc('text', usetex=False)     # turn off TeX for now

fig = plt.figure()           # create a figure object named 'fig'
ax = fig.add_axes([0,0,1,1]) # axes object named 'ax' fills the figure space
                             # numbers in brackets are floats between 0 and 1
                             # [left edge, bottom edge, right edge, and top edge]
                             # with respect to the figure

ax.plot(x, y1)               # plot to axes 'ax'
plt.show()                   # show all plots
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# plot to axes 'ax' with red dashed line with no markers
ax.plot(x, y1, 'r--')

plt.show()
```

`'r--'` in the above `.plot()` command is a shorthand notation for setting the color, linestyle, and/or marker shape. Each is optional and may either be included or omitted. The above example is only used to set the color (red) and linestyle (dashed).

Common color indicators:

- `r` = red
- `g` = green
- `b` = blue
- `c` = cyan
- `m` = magenta
- `y` = yellow 
- `k` = black
- `w` = white

Common linestyle indicators:
- `-` = solid
- `--` = dashed
- `-.` = dash-dot
- `:` = dotted

Common marker indicators:

- `.` = point
- `x` and `X` = $\times$'s
- `+` and `P` = plus signs
- `o` = circle
- `d` and `D` = diamonds
- `s` = square
- `^`, `v`, `<`, and `>` = triangles
- `h` and `H` = hexagons
- `*` = star
- `p` = pentagon
- `8` = octagon

The following code cell creates a "plot" of all of the common marker indicators and linestyles. Matplotlib automatically cycles through different colors for different plots in the same space.

```python
fig = plt.figure(figsize=(6,1))
ax = fig.add_axes([0,0,1,1])
for i,item in enumerate(['-','--','-.',':']):
    ax.plot([0, 17], [2*i+2, 2*i+2], item, lw=2.5)
for i,item in enumerate('.xX+PodDs^v<>hH*p8'):
    ax.plot(i, 0, item, ms=10)
    
ax.set_axis_off()
ax.set_ylim(-1, 9)
plt.show()
```

```python
fig = plt.figure()           
ax = fig.add_axes([0,0,1,1]) 
ax.plot(x, y1, 'r--')
ax.set_title('A Title')      # add a title to 'ax'
ax.set_xlabel('x-axis')      # add an x-axis label to 'ax'
ax.set_ylabel('y-axis')      # add a y-axis label to 'ax'

ymax = np.max(y1[:100])
xmax = x[np.argmax(y1[:100])]

# print a single point by passing x, y values that are single
# values instead of lists or arrays. The above calculation
# assigns the maximum y vlaue in the first half of the plot
# to ymax and the associated x-value to xmax
ax.plot(xmax, ymax, 'b*', markersize=10) 
plt.show()
```

```python
fig = plt.figure()           
ax = fig.add_axes([0,0,1,1]) 
ax.plot(x, y1, 'r--')
ax.set_title('A Title')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.axis(xmin=-2*np.pi, xmax=2*np.pi, ymin=-7, ymax=7) # set display limits of each axis
ax.grid(True)  # turn grid on
plt.show()
```

```python
fig = plt.figure()           
ax = fig.add_axes([0,0,1,1]) 
ax.plot(x, y1, 'r--')
ax.set_title('A Title')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.axis([-2*np.pi, 2*np.pi, -7, 7]) # alternate method; order is xmin, xmax, ymin, ymax
ax.grid(True, ls=':', c='blue')  # turn grid on, linestyle = dotted, color = blue
plt.show()     
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y1, 'r--')
ax.set_title('A Title')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_xlim(-2*np.pi, 2*np.pi) # just set x-axis limits
ax.set_ylim(-7, 7)             # just set y-axis limits
ax.grid(True, ls=':', c='blue')
plt.show()
```

```python
fig, ax = plt.subplots()     # create a figure named 'fig' with one subplot named 'ax'
ax.plot(x, y1, 'r--', lw=2)  # change plot lineweight to 2
ax.set_title('A Title', fontsize=16) # adjust fontsize for title and labels
ax.set_xlabel('x-axis', fontsize=14)
ax.set_ylabel('y-axis', fontsize=14)
ax.set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])  # set locations of x tick marks
ax.set_xticklabels(['-2pi', '-pi', '0', 'pi', '2pi']) # custom labels for x ticks

# can do the above for y-axis using set_yticks and set_yticklabels

ax.set_xlim(-2*np.pi, 2*np.pi)
ax.set_ylim(-7, 7)
ax.grid(True, ls=':', c='blue')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4,3), dpi=150) # custom figure size and dots per inch
ax.plot(x, y1, lw=2, c='darkred', ls='-.')     # alternate method for color and linestyle
ax.set_facecolor('yellow')                     # add a color to the plot area
fig.set_facecolor('grey')                      # add a figure color (around plot axes)
ax.set_title('A Title', fontsize=16, color='blue') # change title or label text color
ax.set_xlabel('x-axis', fontsize=14)
ax.set_ylabel('y-axis', fontsize=14)
ax.set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
ax.set_xticklabels(['-2pi', '-pi', '0', 'pi', '2pi'])
ax.set_xlim(-2*np.pi, 2*np.pi)
ax.set_ylim(-7, 7)
ax.grid(True, ls=':', c='blue')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4,3), dpi=150)
ax.plot(x, y1, lw=2, c='darkred', ls='-.')
ax.set(aspect='equal')  # 1 unit on x-axis equals 1 unit on y-axis

# set all labels at once using set()
ax.set(title='A title', xlabel='x-axis', ylabel='y-axis') 
ax.set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
ax.set_xticklabels(['-2pi', '-pi', '0', 'pi', '2pi'])
ax.set_xlim(-2*np.pi, 2*np.pi)
ax.set_ylim(-7, 7)
ax.grid(True, ls=':', c='blue')
plt.show()
```

### Two plots in the same space

```python
fig, ax = plt.subplots(figsize=(7,3), dpi=150)
ax.plot(x, y1, lw=1.5, c='red', linestyle=(0,(2,1,6,2)), label='$x\cos(x)$')
ax.plot(x, y2, lw=2, c='blue', ls='-.', label='$x\sin(x)$') # label is for legend

# placing $ around items in quotes forces the use of "fancy" text formatting with TeX
# helpful TeX commands include \pi, \sin, \cos, \tan, ^2 (squared), _2 (subscript)
# It may be necessary in some occassions to use a 'raw' string by adding 'r' before
# the quotation mark

ax.set_xlabel('$x$', fontsize=14)
ax.set_ylabel('$y$', fontsize=14)
ax.set_title(r'$x\cos(x)$ and $x\sin(x)$', fontsize=16)
ax.grid(True, linestyle=':', lw=0.5, c='purple')
ax.set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
ax.set_xticklabels(['$-2\pi$', '$-\pi$', '$0$', '$\pi$', '$2\pi$'])
ax.set_xlim([-2*np.pi, 2*np.pi])
ax.set_ylim([-2*np.pi, 2*np.pi])
ax.legend(loc='lower center') # place a legend at the lower center area of 'ax'
# other legend options include: 0 (best), 1 (upper right), 
# 2 (upper left), 3 (lower left), 4 (lower right)
plt.show()
```

### Side-by-side "subplots"

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7,3), dpi=150)
# first two arguments of 'subplots()' set the number of rows
# and number of columns. Can also use nrows=1 and ncols=2
# Subplots with more than one plot axes object will return
# a list instead of a single item. Can use a tuple as the
# second variable before the equal sign like shown here.
# Number of items in tuple must match number of axes created

# 'ax1' refers to the first axes and 'ax2' for the second

ax1.plot(x, y1, lw=1.5, c='red', ls='-')
ax2.plot(x, y2, lw=2, c='blue', ls='-.')
ax1.set_xlabel(r'$x$')
ax2.set_xlabel(r'$x$')
ax1.set_ylabel(r'$y$')
ax2.set_ylabel(r'$y$')
ax1.set_title(r'$x\cos(x)$', fontsize=16)
ax2.set_title(r'$x\sin(x)$', fontsize=16)
ax1.grid(True, linestyle=':', lw=0.5, c='purple')
ax2.grid(True, linestyle='--', lw=0.5)

# TeX is used to make nice looking tick labels with math symbols
ax1.set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
ax1.set_xticklabels([r'$-2\pi$', r'$-\pi$', r'$0$', r'$\pi$', r'$2\pi$'])
ax2.set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
ax2.set_xticklabels([r'$-2\pi$', r'$-\pi$', r'$0$', r'$\pi$', r'$2\pi$'])

ax1.set_xlim([-2*np.pi, 2*np.pi])
ax2.set_xlim([-2*np.pi, 2*np.pi])
ax1.set_ylim([-2*np.pi, 2*np.pi])
ax2.set_ylim([-5, 2])

# 'suptitle' is the title on top of the figure, not a specific axes
# va='baseline' makes this title fit better without any overlap

fig.suptitle(r'$\textrm{\bf Title for both plots}$', fontsize=18, va='baseline')

#using 'tight_layout()' on the figure gives nicer results for subplots
fig.tight_layout()
plt.show()
```

## Using the `.set()` method with axes

```python
fig, ax = plt.subplots(figsize=(4,3), dpi=150)
ax.plot(x, y1, c='green', lw=2, ls=':')
limits=(-2*np.pi,2*np.pi)
ax.set(title="Plot Title", xlabel="x-axis", ylabel="y-axis")
ax.set(aspect="equal", xlim=limits, ylim=limits) # makes the axes fit the data
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4,3), dpi=150)
ax.plot(x, y1, 'o-', c='maroon', lw=4, markevery=0.1, markersize=8)
# circle markers with solid line ('o-'), bigger markers and
# place markers every 0.1 (10%) along x-axis

ax.axis([-2*np.pi, 2*np.pi, -2*np.pi, 2*np.pi])
ax.set(aspect='equal')
ax.set_facecolor('gold')
ax.set_title(r'$\textrm{\bf Bold, Serif Title}$', fontsize=18)
ax.grid(c='maroon', ls=':')
plt.show()
```

## Polar plotting using object-oriented commands

```python
fig = plt.figure(figsize=(8,4))
# create the figure using 'figure()' for polar plots
# use polar=True in 'add_axes()' to create a polar plot
# then just use 'plot()' as usual except (theta, r) not (x, y)

ax = fig.add_axes([0,0,0.5,1], polar=True) # 'ax' uses left half of figure space
ax.plot(x, y1, 'r', lw=1.5)
ax.set_title('$x\cos(x)$', fontsize=16)
ax.set_rlim(-2*np.pi, 2*np.pi)
ax.set_xticks([])  # 'x' is the same as 'theta' for polar plots
                   # the empty list makes is so there are not ticks

ax.set_yticks([-2*np.pi, 0, 2*np.pi]) # 'y' is 'r' for polar
ax.set_yticklabels(['$-2\pi$', '$0$', '$2\pi$'], fontsize=12)

ax2 = fig.add_axes([0.5, 0, 1, 1], polar=True) # 'ax2' uses right half of figure space
ax2.plot(x, y2, 'b', lw=1.5)
ax2.set_title('$x\sin(x)$', fontsize=16)
plt.show()
```

## MATLAB-style non-object-oriented plotting

Similar to the object-oriented plotting commands, Matplotlib's MATLAB-style state-machine plotting can generate the same results. There are limits to the control available to "tweak" plots and creating certain types of plots or plot arrangements when using this approach.

### Two plots in the same space

```python
plt.figure(figsize = (7,3), dpi=150)
plt.plot(x, y1, 'r-', lw=1.5, label='$x\cos(x)$')
plt.plot(x, y2, 'b-.', lw=2, label='$x\sin(x)$')
plt.legend(loc="lower center")
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$x\cos(x)$ and $x\sin(x)$', fontsize=16)
plt.grid(True, linestyle=':', lw=0.5, c='purple')
plt.axis([-2*np.pi, 2*np.pi, -2*np.pi, 2*np.pi])
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],['$-2\pi$', '$-\pi$', '$0$', '$\pi$', '$2\pi$'])
plt.show()
```

### Side-by-side "subplots"

```python
plt.figure(figsize = (7,3), dpi=150)
plt.subplot(121)
plt.plot(x, y1, 'r-', lw=1.5)
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$x\cos(x)$', fontsize=16)
plt.grid(True, linestyle=':', lw=0.5, c='purple')
plt.axis([-2*np.pi, 2*np.pi, -2*np.pi, 2*np.pi])
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],['$-2\pi$', '$-\pi$', '$0$', '$\pi$', '$2\pi$'])
plt.subplot(122)
plt.plot(x, y2, 'b-.', lw=2)
plt.grid(True, linestyle='--', lw=0.5)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$x\sin(x)$', fontsize=16)
plt.axis([-2*np.pi, 2*np.pi, -5, 2])
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],['$-2\pi$', '$-\pi$', '$0$', '$\pi$', '$2\pi$'])
plt.suptitle(r'$\textrm{\bf Title for both plots}$', fontsize='18', va='baseline')
plt.tight_layout()
plt.show()
```
